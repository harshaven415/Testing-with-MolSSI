```python
#One common way to analyze tabular data is to graph it

#From the previous lesson, we had a variable in distances

import os
import numpy

distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter = ',', dtype = 'unicode')

headers = distances[0]
data = distances[1:]
data = data.astype(float)



import matplotlib.pyplot

matplotlib.pyplot.figure() #This will initailize a new figure
matplotlib.pyplot.plot(data[:,1])
```

```python
#We can shorten the function quite easily using this technique:

import matplotlib.pyplot as plt

plt.figure()
plt.plot(data[:,1])
```

```python
#Now, let us learn how to label plots and save figures

sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')

fig_1 = plt.plot(data[:,1], label = sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python
#Now, let us make this aesthetic as Glen said, so let's increase resolution

sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')

fig_1 = plt.plot(data[:,1], label = sample)
plt.legend()

plt.savefig(F'{sample}.png', dpi = 300)
```

```python
#We can plot more than one data set on the same graph by using another plot command

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')
plt.plot(data[:,1], label = headers[1])
plt.plot(data[:,2], label = headers[2])
plt.legend()

plt.savefig('two_samples.png')
```
