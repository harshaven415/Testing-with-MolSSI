```python
for f in filenames:
    #We are now getting the file names
    file_name = os.path.basename(f)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]
    
    #Now we are reading the data
    outfile = open(f, 'r')
    data = outfile.readlines()
    outfile.close()
    
    #Now we loop through the data
    for line in data:
        if 'Final Energy' in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            print(molecule_name, energy)
```

```python
#w means open the file for writing, and the + means create a file
#a will append to an existing file
#while w+ overwrites to make a new file, a+ will not, only append

datafile = open('energies.txt','w+')  #This opens the file for writing
for f in filenames:
    # Get the molecule name
    file_name = os.path.basename(f)
    split_filename = file_name.split('.')
    molecule_name = split_filename[0]

    # Read the data
    outfile = open(f,'r')
    data = outfile.readlines()
    outfile.close()

    # Loop through the data
    for line in data:
        if 'Final Energy' in line:
            energy_line = line
            words = energy_line.split()
            energy = float(words[3])
            datafile.write(F'{molecule_name} \t {energy} \n')
datafile.close()
```

```python
import numpy
import os
file_location = os.path.join('data', 'water.xyz')
xyzfile = open(file_location,"r")
data=xyzfile.readlines()
num_atoms = int(data[0])
data = data[2:]

symbols = []
coordinates = []

for atom in data:
    atom_data = atom.split()
    symbol = atom_data[0]
    symbols.append(symbol)
    x, y, z = atom_data[1], atom_data[2], atom_data[3]
    coordinates.append([float(x), float(y), float(z)])

for num1 in range(0,num_atoms):
    for num2 in range(0,num_atoms):
        atom1 = coordinates[num1]
        atom2 = coordinates[num2]
        x_distance = atom1[0] - atom2[0]
        y_distance = atom1[1] - atom2[1]
        z_distance = atom1[2] - atom2[2]
        bond_length_12 = numpy.sqrt(x_distance**2+y_distance**2+z_distance**2)
        
        
if bond_length_12 > 0 and bond_length_12 <= 1.5:
        print(F'{symbols[num1]} to {symbols[num2]} : {bond_length_12:.3f}')          
```

```python
import numpy
import os

file_location = os.path.join('data', 'water.xyz')
xyz_file = numpy.genfromtxt(fname=file_location, skip_header=2, dtype='unicode')
symbols = xyz_file[:,0]
coordinates = (xyz_file[:,1:])
coordinates = coordinates.astype(float)
num_atoms = len(symbols)

BLfile = open('bond_lengths.txt','w+')
for num1 in range(0,num_atoms):
    for num2 in range(0,num_atoms):
        if num1<num2:
            x_distance = coordinates[num1,0] - coordinates[num2,0]
            y_distance = coordinates[num1,1] - coordinates[num2,1]
            z_distance = coordinates[num1,2] - coordinates[num2,2]
            bond_length_12 = numpy.sqrt(x_distance**2+y_distance**2+z_distance**2)
            if bond_length_12 > 0 and bond_length_12 <= 1.5:
                BLfile.write(F'{symbols[num1]} to {symbols[num2]} : {bond_length_12:.3f}')
                
BLfile.close()
```

```python
#Now, we will plot and visualize data using data and matplotlib

#Let's import the data and turn everything into a float 
import os
import numpy

distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0]
data = distances[1:]
data = data.astype(float)

#Now, we import the python library matplotlib

import matplotlib.pyplot

matplotlib.pyplot.figure() #This will create a new figure
matplotlib.pyplot.plot(data[:,1])
```

```python
#matplotlib.pyplot is a often shortened to plt using the following

import matplotlib.pyplot as plt

plt.figure()
plt.plot(data[:,1])
```

```python
#Now we will try labeling plots and saving figures 

sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
fig_1 = plt.plot(data[:,1], label = sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python
#We often need to increase image quality or resolution
#This is done with the dpi command in plt.savefig

sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')
fig_1 = plt.plot(data[:,1], label = sample)
plt.legend()

plt.savefig(F'{sample}.png', dpi = 300)
```

```python
#Now, we will try overlaying multiple sets of data into a figure

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')

plt.plot(data[:,1], label = headers[1])
plt.plot(data[:,2], label = headers[2])
plt.legend()

plt.savefig('TwoSamples.png')
```

```python
#Now, let us include all samples on the same plot using a for loop

for col in range(1, len(headers)):
    fig = plt.plot(data[:, col], label=headers[col])
    plt.legend()
    
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Angstrom)')
plt.savefig('All_Samples.png', dpi = 300)
```

```python
#To make a different plot for each sample move the plt.figure()

for col in range(1, len(headers)):
    plt.figure()
    sample = headers[col]
    
    plt.plot(data[:, col], label = sample)
    plt.legend()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (Angstrom)')
    
    plt.savefig(F'{sample}.png')
```

```python
#Now we will normalize the data using minimum and maximum values

data_min = numpy.min(data[:, 1:]) * 0.9
data_max = numpy.max(data[:, 1:]) * 1.1

for col in range(1, len(headers)):
    plt.figure()
    sample = headers[col]
    
    plt.plot(data[:, col], label = sample)
    plt.ylim(data_min, data_max)
    plt.legend()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (Angstrom)')
    
    plt.savefig(F'{sample}.png')
```

```python
#Let us revisit the geometry analysis

import numpy
import os

file_location = os.path.join('data', 'water.xyz')
xyz_file = numpy.genfromtxt(fname=file_location, skip_header=2, dtype = 'unicode')

symbols = xyz_file[:, 0]
coordinates = xyz_file[:, 1:]
coordinates = coordinates.astype(float)

num_atoms = len(symbols)
for num1 in range(0, num_atoms):
    for num2 in range(0, num_atoms):
        if num1 < num2:
            x_distance = coordinates[num1, 0] - coordinates[num2,0]
            y_distance = coordinates[num1, 1] - coordinates[num2,1]
            z_distance = coordinates[num1, 2] - coordinates[num2,2]
            b_length_12 = numpy.sqrt(x_distance**2 + y_distance**2 + z_distance**2)
            if b_length_12 > 0 and b_length_12 <=1.5:
                print(F'{symbols[num1]} to {symbols[num2]} : {b_length_12:.3f}')
```

```python
#Clearly, the above code is very wordy and not fun to read or type
#So let us simplify the process by using a function

def calculate_distance(atom1_coord, atom2_coord):
    x_distance = atom1_coord[0] - atom2_coord[0]
    y_distance = atom1_coord[1] - atom2_coord[1]
    z_distance = atom1_coord[2] - atom2_coord[2]
    b_length_12 = numpy.sqrt(x_distance**2 + y_distance**2 + z_distance**2)
    return b_length_12
```

```python
for num1 in range(0, num_atoms):
    for num2 in range(0, num_atoms):
        if num1 < num2:
            b_length_12 = calculate_distance(coordinates[num1], coordinates[num2])
            if b_length_12 > 0 and b_length_12 <=1.5:
                print(F'{symbols[num1]} to {symbols[num2]} : {b_length_12:.3f}')
```
